Add a new column called Total using pandas and store total of sub1,sub2 and sub3

```python
result['Total'] = result[['sub1', 'sub2','sub3']].sum(axis=1)
```

```python
result
```

Add a new column called Average. Store average value of sub1,sub2,sub3

```python
result['Average'] = result[['sub1', 'sub2','sub3']].mean(axis=1)
```
